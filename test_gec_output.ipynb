```python
def CheckSpace(original, compare):
    ori_num =[]
    com_num = []
            
    corr = []
    err = []
    for i in range(len(original)):
        original_sentence = original[i].strip().lower().strip().rstrip(',.')
        compare_sentence = compare[i].strip().lower().strip().rstrip(',.')
        
        ori_num.append(len(original_sentence.split(' ')))# word count in sentence
        com_num.append(len(compare_sentence.split(' ')))
        
    for j in range(len(original)):
        if ori_num[j] == com_num[j]:
            corr.append(j)
        else:
            err.append(j)
    #        print(trans_num[i],trans[i])
    #        print(asr_num[i],asr[i])
    print("띄어쓰기 문제가 있는 문장: ",len(err))
    print("띄어쓰기 문제가 없는 문장: ",len(corr))
```

```python
def CheckWER(original, compare):
    ori2com=[]
    for i in range(len(original)):
        original_sentence = original[i].strip().lower().strip().rstrip(',.')
        compare_sentence = compare[i].strip().lower().strip().rstrip(',.')
     #   print(original_sentence,'\n', compare_sentence,'\n')
        error = wer(original_sentence,compare_sentence)
        ori2com.append(error)

    print("총 문장:",len(ori2com))
    print("\nWER: {:.5f}".format(sum(ori2com)/len(ori2com)))
    print("Accuracy: {:.5f}".format(100-sum(ori2com)/len(ori2com)))

    err = 0
    corr = 0
    for i in range(len(ori2com)):
        if ori2com[i] == 0:
            corr += 1
        else:
          #  print(i) -오류 없는 문장 index
            err += 1

    print("\n오류가 있는 문장:",err)
    print("오류가 없는 문장:",corr)
```

# Sound Hamming Distance로 문장 고쳐보기

문법 오류 교정 결과에다가 sound hamming distance 적용해보기

```python
from eudex import eudex
e = eudex
def eudexDistance(w1,w2):
    result = sum(1 for _ in bin(e(w1) ^ e(w2)) if _ == '1') # number of one after xoring hashes
    return result
```

```python
def EditWord(original, hypothesis):
    result_list = []
    for i, j in zip(original, hypothesis):
        result = []
        i_tokens = list(map(str,i.lower().strip().rstrip(',.').split()))
        j_tokens = list(map(str,j.lower().strip().rstrip(',.').split()))
        if len(i_tokens) == len(j_tokens):
            if i_tokens == j_tokens:
                result = i_tokens
            else: 
                for i_tok, j_tok in zip(i_tokens, j_tokens):
                    if eudexDistance(i_tok, j_tok) <= 1: # sound hd가 2이하면 교정한 토큰 붙임.
                        result.append(j_tok) 
                    else: 
                        result.append(i_tok) # 아니면(2보다 크면) 원래 음성인식 토큰을 붙임     
        else:
            result = i_tokens
        result = ' '.join(result) # attach tokens to sentence
        result_list.append(result) # processed sentence
    
    return result_list 
```

```python
def EditSentence(transcript, original, hypothesis):
    result_list = []
    count = 0
    right = 0
    for idx,(i, j) in enumerate(zip(original, hypothesis)):
        i = i.lower().strip().rstrip(',.')
        j = j.lower().strip().rstrip(',.')
        if i == j:
            result_list.append(i)
        else:
            if eudexDistance(i, j) <= 1:  # sound hd가 1이하면 교정한 sentence 붙임.
                result_list.append(j)
                count += 1
                ori = transcript[idx].strip().lower().strip().rstrip(',.')
                if wer(ori, j) == 0:
                    right += 1
            else:
                result_list.append(i) # 아니면(2보다 크면) 원래 음성인식 sentence 붙임 
    print(right,'/',count)
        
    return result_list 
```
